```python
import pandas as pd

# Assuming 'df' is your DataFrame that contains the ETF data including the date column

# Convert all columns except the date to numeric, coercing errors to NaN
for column in df.columns[1:]:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Now, calculate the month-on-month percentage change, skipping the first row
monthly_returns = df.iloc[1:, 1:].pct_change()

# Calculate the average monthly return for each ETF, excluding the date column
expected_returns = monthly_returns.mean() * 100  # Convert to percentage

# Display the expected returns
print(expected_returns)
```

```python
import numpy as np

for column in df.columns[1:]:
    df[column] = pd.to_numeric(df[column], errors='coerce')
# Calculate the mean monthly return for each ETF
mean_monthly_returns = df.iloc[1:].pct_change().mean()

# Calculate the standard deviation of the monthly return for each ETF
std_monthly_returns = df.iloc[1:].pct_change().std()

# Annualize the mean returns by multiplying by 12 (for 12 months)
annual_returns = mean_monthly_returns * 12

# Annualize the standard deviation by multiplying by the square root of 12 (for 12 months)
std_annual_returns = std_monthly_returns * np.sqrt(12)

# Generate the correlation table of the monthly returns for each ETF
correlation_table = df.iloc[1:].pct_change().corr()

# Generate the covariance table of the monthly returns for each ETF
covariance_table = df.iloc[1:].pct_change().cov()

# Output the results
mean_monthly_returns, std_monthly_returns, annual_returns, std_annual_returns, correlation_table, covariance_table
```

```python
import pandas as pd
from scipy.optimize import minimize

# Assuming `df` is your DataFrame with the data
# Ensure you've replaced this with the actual code to load your data into a DataFrame
# For example: df = pd.read_csv('your_data.csv')

# Define the objective function (we choose variance here for demonstration)
def portfolio_variance(weights, covariance_matrix):
    return weights.dot(covariance_matrix).dot(weights.T)

# Define the constraint functions
def constraint_expected_return(weights, expected_returns, target_return):
    return weights.dot(expected_returns) - target_return

def bounds(weights):
    return [(0, 1) for _ in weights]  # weights between 0% and 100%

# The initial guess for the weights
initial_weights = df['Weight'].values
# The expected returns for each asset
expected_returns = df['E [return]'].values
# Calculate the covariance matrix for the 'Risk' column
# Normally you would use historical returns to calculate the covariance matrix
# Here we just square the 'Risk' as a placeholder
covariance_matrix = pd.DataFrame(df['Risk']).pow(2)

# The target return
target_return = 0.05  # This is just an example target return

# The constraints dictionary
constraints = [
    {'type': 'eq', 'fun': lambda weights: sum(weights) - 1},  # The sum of weights is 1
    {'type': 'eq', 'fun': lambda weights: constraint_expected_return(weights, expected_returns, target_return)}
]

# Perform the optimization
result = minimize(portfolio_variance, initial_weights, args=(covariance_matrix,), method='SLSQP', bounds=bounds(initial_weights), constraints=constraints)

if result.success:
    optimized_weights = result.x
    print('Optimized weights:', optimized_weights)
else:
    print('Optimization failed:', result.message)
```
