```python
import yfinance as yf
import pandas as pd

# Define the symbols for the assets you're interested in
symbols = {
    'S&P 500': '^GSPC',  # S&P 500 index
    'Gold': 'GLD',  # SPDR Gold Shares ETF as a proxy for XAU Gold
    'Thai Bond': 'THD',  # Using iShares MSCI Thailand ETF as a placeholder (Note: This is not a bond ETF)
    'US Bond': 'TLT'  # iShares 20+ Year Treasury Bond ETF
}

# Define the start and end dates for the historical data
start_date = '2010-01-01'
end_date = '2023-01-01'

# Create an empty dictionary to store the data
data = {}

# Download the data
for name, symbol in symbols.items():
    data[name] = yf.download(symbol, start=start_date, end=end_date)

# Accessing the data
# For example, to print the data for Gold
print("Gold (GLD) Historical Data:")
print(data['Gold'].head())

# Remember to replace 'PROPERTY_SYMBOL.BK' with the actual symbol of the Thai property stock you're interested in.
```

```python
data['S&P 500']['Close'].plot()
```

```python
# loop data['gold']['adj close'] and perform pct_change()
# store the result in a new column in the dataframe
data['Gold']['Return'] = data['Gold']['Adj Close'].pct_change()
# make new pd dataframe of return year by year
gold_return = data['Gold']['Return'].resample('Y').sum()
gold_return = gold_return.to_frame()


```

```python

# gold_return to csv
gold_return.to_csv('gold_return.csv')
```

```python
# cumulative return
gold_cum_return = (1 + gold_return).cumprod() - 1
# annualized return
gold_annual_return = (1 + gold_return.mean()) ** 12 - 1
    
print(gold_cum_return, gold_annual_return)
```
